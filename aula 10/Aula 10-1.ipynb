# Introdução a Matplotlib

## Importando Matplotlib e NumPy

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Inserindo vertices

```python
x = np.array([0,0,0,1,1,2,3,3,3,3,4,6,6,8,10])
y = np.array([5,10,10,5,5,15,15,20,20,0,25,0,10,25,10])
```

## Plotagem e visualização

```python
plt.plot(x, y,'o')
plt.show()
```
